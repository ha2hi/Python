```python
### 1.1 데이터 파이프라인 소개
- 테스크간의 의존성을 명확하게 확인하는 방법 중 하나는 데이터 파이프라인을 그래프로 표현한는 것입니다.
- DAG란 방향성 비순환 그래프로 파이프라인의 순서대로 작동하나 비순환 그래프이기 때문에 서로 의존적이면 안된다.

### 1.2 파이프라인 그래프 실행 순서
1. 이전 테스크가 완료되었는지 확인합니다.
2. 완료되었다면 다음 실행할 태스크를 대기열에 추가합니다.
3. 대기열에 있는 태스크를 실행하고 완료되면 완료 표시를 합니다.
4. 모든 태스크가 완료될 때까지 1단계로 돌아갑니다.

### 1.3 그래프 파이프라인 장점
1. 각각의 태스크를 병렬로 실행할 수 있다.
2. 실패한 태스크 혹은 그 이후 태스크만 실행할 수 있다.

```

```python
### 1.2 Airflow 소개
### 1.2.2 파이프라인 스케줄링 및 실행
- Airflow 스케줄러
    - DAG를 분석하고 현재 시점에서 DAG의 스케줄이 지난 경우 Airflow 워커에 DAG 태스크를 예약합니다.
- Airflow 워커
    - 예약된 태스크를 선택하고 실행합니다.
- Airflow 웹서버
    - 스케줄러에서 분석한 DAG를 시각화하고 DAG 실행과 결과를 확인할 수 있는 주요 인터페이스를 제공합니다.
```

```python
### 1.3.1 Airflow를 선택하는 이유
- 다양한 시스템과 통합이 가능합니다.
- 전체 파이프라인을 재실행할 필요 없는 효율적인 파이프라인 구축이 가능합니다.
- 백필 기능을 사용하면 과거 데이터를 손쉽게 재처리할 수 있습니다.
- 파이프라인 실행결과를 모니터링할 수 있고 오류 디버깅하기 위한 편리한 뷰를 제공합니다.

### 1.3.2 Airflow가 적합하지 않은 경우
- 스트리밍 워크플로에 적합하지 않습니다.
- 추가 및 삭제 태스크가 빈번한 동적 파이프라인의 경우 적합하지 않을 수 있습니다.
```
