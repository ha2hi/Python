```python
# https://www.acmicpc.net/problem/17780
import sys
def move(num):
    x, y, d = chess[num]
    if chess_map[x][y][0] != num:
        return 0
    
    xx = x + dx[d]
    yy = y + dy[d]
    if not(0<=xx<n and 0<=yy<n) or board[xx][yy] == 2:
        if d == 0 or d== 1:
            dd = (d+1)%2
        else:
            dd = (d-1)%2 + 2
        chess[num][2] = dd
        xx = x + dx[dd]
        yy = y + dy[dd]
        if not(0<=xx<n and 0<=yy<n) or board[xx][yy] == 2:
            return 0
    
    chess_set = chess_map[x][y]
    chess_map[x][y] = []

    if board[xx][yy] == 1:
        chess_set = chess_set[::-1]
    
    for i in chess_set:
        chess_map[xx][yy].append(i)
        chess[i][:2] = [xx,yy]
    
    if len(chess_map[xx][yy]) >= 4:
        return 1
    return 0


n, k = map(int, input().split())
board = [list(map(int, input().split())) for _ in range(n)]
chess_map = [[[] for _ in range(n)] for _ in range(n)]
chess = [0 for _ in range(k)]
dx = [0,0,-1,1]
dy = [1,-1,0,0]

for i in range(k):
    x,y,d = map(int, input().split())
    chess_map[x-1][y-1].append(i)
    chess[i] = [x-1, y-1, d-1]

days = 1
while days <= 1000:
    for i in range(k):
        flag = move(i)
        if flag:
            print(days)
            sys.exit()
    days += 1
print(-1)
```
